```
def calcu(area, vTe):
  return (area/vTe)*c

print("1 – Medicion de dos terrenos \nAl medir dos lotes o terrenos, se dice que hay aproximadamente 1 m de \nerror en la medición. Un terreno es de 300 metros y otro de 2000.\n")
#lote 300 metros
c = 100
vMetro=1
vTerreno=300
print("Ep = (", vMetro , "/", vTerreno,")*",c )
print("EP = ",calcu(vMetro,vTerreno))

#lote 200 metros
vTrreno2=2000
print("Ep = (", vMetro , "/", vTrreno2,")*",c )
print("EP = ",calcu(vMetro,vTrreno2),"%")

print("\nEn un laboratorio es entregado un bloque de aluminio. Al medir las dimensiones del bloque y calcular su masa y volumen,\nse determina la densidad del mismo (2,68 g/cm3).Sin embargo, al revisar la tabla numérica del material, ésta indica \nque la densidad del aluminio es de 2,7 g/cm3. De esta manera, el error absoluto y porcentual se calcularía de la siguiente forma:\n")
vMasa=2.68
vDencidad=2.7
print("EA =",vMasa,"-",vDencidad)
Ea = vDencidad-vMasa
print("Ea = ",Ea)
print("Ep = (", Ea , "/", vDencidad,")*",c )
print("EP = ",calcu(Ea,vDencidad),"%")

print("\nSe asumio que 1,000,000 de personas irian a un evento determinado. Sin embargo, el numero exacto de personas \nque fue a dicho evento fue de 88,000. El error absoluto y porcentual serian los siguientes:")
vPersonas=1000000
vPerA=88000
print("EA =",vPersonas,"-",vPerA)
Ea = vPersonas-vPerA
print("Ea = ",Ea)
print("Ep = (", Ea , "/", vPersonas,")*",c )
print("EP = ",calcu(Ea,vPersonas),"%")

print("\nEl tiempo que se calcula debe tardar una pelota en alcanzar el suelo después de ser lanzada a una distancia de 4 metros, \nes de 3 segundos. Sin embargo, en el momento de la experimentación, se descubre que la pelota tardó 2,1 segundo el alcanzar el suelo.\n")
vPelota=2.1
vCalPelota=3
print("EA =",vPelota,"-",vCalPelota)
Ea = vCalPelota-vPelota
print("Ea = ",Ea)
print("Ep = (", Ea , "/", vPelota,")*",c )
print("EP = ",calcu(Ea,vPelota),"%")

print("\nSe aproxima que si un auto va a 60 km, éste llegará a su destino en 1 hora. No obstante, en la vida real, el auto tardó 1,2 horas en llegar a su destino. El error porcentual de este cálculo de tiempo se expresaría de la siguiente forma:\n")
vHora=1
vHoAuto=1.2
print("EA =",vHora,"-",vHoAuto)
Ea = vHora-vHoAuto
print("Ea = ",Ea)
print("Ep = (", Ea , "/", vHoAuto,")*",c )
print("EP = ",calcu(Ea,vHoAuto),"%")

print("\nSe mide una longitud cualquier por un valor de 30 cm. Al verificar la medición de esta longitud se evidencia que hubo un error de 0,2 cm. El error porcentual en este caso se manifestaría de la siguiente forma:\n")
vLongi=30
vLonMed=0.2
print("Ep = (", vLonMed , "/", vLongi,")*",c )
print("EP = ",calcu(vLonMed,vLongi))

print("\nEl cálculo de longitud de un puente según sus planos es de 100 m. Sin embargo, al confirmar dicha longitud una vez es construido se evidencia que en realidad es de 99,8 m de largo. El error porcentual se evidenciaría de esta forma.\n")
vPuenPla=100
vPuenCon=99.8
print("EA =",vPuenPla,"-",vPuenCon)
Ea = vPuenPla-vPuenCon
print("Ea = ",Ea)
print("Ep = (", Ea , "/", vPuenCon,")*",c )
print("EP = ",calcu(Ea,vPuenCon),"%")

print("\nLa cabeza de un tornillo fabricado de forma estándar es dada para tener 1 cm de diámetro. Sin embargo, a la hora de medir este diámetro, \nse observa que la cabeza del tornillo tiene en realidad 0.85 cm. El error porcentual sería?\n")
vTornEst=1
vTornReal=0.85
print("EA =",vTornEst,"-",vTornReal)
Ea = vTornEst-vTornReal
print("Ea = ",Ea)
print("Ep = (", Ea , "/", vTornReal,")*",c )
print("EP = ",calcu(Ea,vTornReal),"%")

print("\nDe acuerdo a su volumen y materiales, se calcula que el peso de un objeto dado es de 30 kilos. Una vez el objeto es analizado, se observa que su peso real es de 32 kilos.\n")
vPesoObj=30
vPesoReal=32
print("EA =",vPesoReal,"-",vPesoObj)
Ea = vPesoReal-vPesoObj
print("Ea = ",Ea)
print("Ep = (", Ea , "/", vPesoReal,")*",c )
print("EP = ",calcu(Ea,vPesoReal),"%")

print("\nEn un laboratorio es estudiada una lámina de acero. Al medir las dimensiones de la lámina y calcular su masa y volumen, se determina la densidad de la misma (3,51 g/cm3). Sin embargo, al revisar la tabla numérica del material, ésta indica que la densidad del acero es de 2,85 g/cm3. De esta manera, el error absoluto y porcentual se calcularía?\n")
vMasaObj=3.51
vMasaDen=2.85
print("EA =",vMasaObj,"-",vMasaDen)
Ea = vMasaObj-vMasaDen
print("Ea = ",Ea)
print("Ep = (", Ea , "/", vMasaDen,")*",c )
print("EP = ",calcu(Ea,vMasaDen),"%")


```
